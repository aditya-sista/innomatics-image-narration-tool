```python
from transformers import pipeline
from PIL import Image

classifier = pipeline(
    task="object-detection",
    model="facebook/detr-resnet-50"
)
```

```python
image = Image.open("images/crossing.jpg")
```

```python
detected_objects = classifier(image)
```

```python
! uv pip install collections
```

```python
from collections import Counter

labels = [obj["label"] for obj in detected_objects]
total_objects = Counter(labels)
```

```python
def detected_objs_to_text(total_objects):
    if not total_objects:
        return "There are no detected objects in the image."

    parts = []
    for label, count in total_objects.items():
        if count == 1:
            parts.append(f"One {label}")
        else:
            parts.append(f"{count} {label}s")

    if len(parts) == 1:
        return f"There is {parts[0]}."
    else:
        return "There are " + ", ".join(parts[:-1]) + " and " + parts[-1] + "."

description = detected_objs_to_text(total_objects)
```

```python
! uv pip install soundfile
```

```python
from transformers import pipeline
import soundfile as sf

classifier = pipeline(
    task="text-to-speech",
    model="suno/bark-small"
)

speech = classifier(description)

sf.write("hf_output.wav", speech["audio"], speech["sampling_rate"])
```

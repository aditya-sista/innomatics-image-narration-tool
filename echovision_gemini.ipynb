```python
from google import genai
from google.genai import types

f = open("keys/.gemini.txt")
key = f.read()

client = genai.Client(api_key=key)

with open("images/crossing.jpg", "rb") as f:
    image_bytes = f.read()

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=[
        types.Part.from_bytes(data=image_bytes, mime_type="image/jpeg"), 'Briefly describe what is in this image.'
    ]
)

description = response.text
print(description)
```

```python
! uv pip install soundfile
```

```python
from transformers import pipeline
import soundfile as sf

classifier = pipeline(
    task="text-to-speech",
    model="suno/bark-small"
)

speech = classifier(description)

sf.write("gemini_output.wav", speech["audio"], speech["sampling_rate"])
```
